### 1.2 Zillow Business Background:

#### 1.2.1 What is Zillow Zestimate?

First of all, let us understand what Zillow’s Zestimate actually is. Zestimate is a valuation tool which can be used by anyone to get a prediction about the approximate value of real estate properties. This tool can be used to predict the value of an owned property that someone wants to sell or serve as a benchmark tool for buyers who want to buy a property in a certain residential area. Therefore, Zestimate provides significant price and information transparency to consumers that was previously only accessible to brokers. It provides data about more than 100 million homes across the United States.

#### 1.2.2 Bid-ask Spread in the Housing Market

In a commodity market, there is always a bid-ask spread because there is a difference in the price that is asked for the commodity and the price that is bid for it. The valuation tool steps in here and informs each user about the estimated price, which in theory, should reduce the bid-ask spread between both parties. It should help to prevent buyers from buying a property that was a lot less valuable. And it should prevent sellers from selling a house for a fraction of what it is worth. This higher information transparency should reduce the adverse selection that is prevalent in the real estate market.

#### 1.2.3 Zillow’s Business Model with Zestimate

But what is Zillow’s business intention behind providing such a service?

The reasons are manifold: The valuation tool is a strong piece in the puzzle of creating traffic on their online marketplace to get stakeholders (sellers, buyers, and professional brokers) on the platform. It is an effective “foot in the door” to get customers who will use further provided services (listings, broker contacts, all-in-one solutions) which generate revenue and further bind the customer to the platform, hopefully maximizing the lifetime value of each user. Zillow provides a marketplace for sellers, buyers and professional brokers, and their additional services for free or priced help them to dominate the online market for real estate in the USA.

However, Zillow also used Zestimate strategically to use the housing valuation data and estimations to engage as a buyer and seller in the market. In simple language: they would buy properties and then try to sell it for a higher price, coined as a “flipping business”. Then they would analyze the best prospects, which then got bought, maybe renovated, and then they tried to sell them for a higher price. In general, Zillow was able to anticipate the price of housing property with a 14 median absolute percent error using Zestimate. Zillow even improved the Zestimate model and lowered its error rate to 5%. Yet, Zillow had to stop its flipping business, had to cut its workforce by 25% (fire 2000 employees) and Zillow Offers (the company branch responsible for buying & selling houses) made a $420 million dollar loss in 2021.

#### 1.2.4 Why Zillow’s Prediction Model failed

For us as a data science team it is really interesting to explore and understand why Zillow’s prediction model did not work out. The crucial question is, which factors could play a role that influence the accuracy of a housing pricing model as well as which factors are maybe not considered by the model, but are in fact influencing house prices? This is the question we want to answer with our analysis.

After viewing the data set it becomes clear that several factors which are fair to assume to influence housing prices are not represented in the data. External factors (i.e. Covid-19, inflation, politics, local demographics, climate change) are not considered as data in Zillows model. The same accounts for buyer behavior and preferences, or the buying/selling supply and demand in the same area. Also, the housing data comes from private users who input their housing data, which might be subject to inconsistent levels of quality as not everyone might be truthful about the characteristics of their home for whatever reason. The majority of users probably answer truthfully, as they have the right incentive to do so. This does not necessarily need to hinder the model to predict the prices accurately, but it underlines that many factors influence the pricing of a particular parcel showing the complexity of the task overall as a model is naturally a simplification of reality. 

#### 1.2.5 The Zollow Data Science Competition

From the perspective of Zillow, a machine learning model that could accurately predict which houses are going to be sold would most certainly be the “holy grail” that would allow them to engage again in the house “flipping” business they stopped due to the inaccuracy of their model and the complexity of the task. But it is no surprise that there is a competitive Data Science Competition on Kaggle with a Prize Money of staggering $1,200,000 to improve the accuracy of the Zestimate even further. 

#### 1.2.6 Predicting Log Error

As mentioned, we do not have the data to build a model that predicts the house price or if a house will be sold, because the data only tells us the log-error between the previous Zillow estimate model and the actual selling price for a sample of the properties listed in a given time interval. But we can predict the log-error and compare it to the log-error provided in the data set. In essence, the more accurately we predict the log-error of the model, the better we will understand the shortcomings of Zillow’s model. The deviation in log-error will allow us to identify which features are responsible for a higher log-error, allowing us to optimize the model for a lower log-error in general to be implemented by Zillow.

#### 1.2.7 logerror = log⁡(Zestimate) - log⁡(SalePrice)

In general, a logarithmic error is one where the error is proportional to the logarithm of the predicted value. This can be useful when working with data sets where the values can vary widely (e.g. prices of houses). The logarithmic error is a measure of how well a prediction model predicts the house price of a property. A low logarithmic error indicates that the model is able to predict house prices well. A higher logarithmic error may be indicative of a higher house price.

### 6.2 Final Preprocessing Pipeline

```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
    
numcols.remove('logerror')
num_features = numcols
num_pipeline = Pipeline([('selector', DataFrameSelector(num_features)),
                         ('imputer', SimpleImputer(strategy="median")),
                         ('scaler', StandardScaler())])

cat_features = catcols
cat_pipeline = Pipeline([('selector', DataFrameSelector(cat_features)),
                         ('cat_encoder', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(transformers=[('num_features', num_pipeline, num_features), ('cat_features', cat_pipeline, cat_features)])
```

## 8. Hyperparameter Tuning & Feature Importance using RandomForest

#### 8.1 Hyperparameter Tuning

To improve upon our best performing model (RandomForest), we tune the hyperparameters using GridSearchCV.

```python
rf = RandomForestRegressor()

parameters_rf = {'n_estimators': [100, 500, 1000], 'max_depth': [ 1, 3, 5]}

rf_gscv = GridSearchCV(estimator=rf, param_grid=parameters_rf, cv=5, scoring='neg_mean_absolute_error')
rf_gscv.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

print('Best parameters:', rf_gscv.best_params_)
print('Mean cross-validated MAE score of the best_estimator:', abs(rf_gscv.best_score_))
```
